```python
obesity = pd.read_csv('obesity.csv')
obesity.info()
```

Very complete data, no nulls in any of the columns

```python
obesity = obesity.rename({
    'family_history_with_overweight': 'Overweight_Family_History', 
    'FAVC': 'High_Calorie_Consumption_Often',
    'FCVC': 'Vegetable_Consumption_Often',
    'NCP': 'Num_Main_Meals_Daily',
    'CAEC': 'Eat_Between_Meals',
    'SMOKE': 'Smoke_Regularly',
    'CH2O': 'Water_Drank_Daily',
    'SCC': 'Calories_Monitored_Daily',
    'FAF': 'Workout_Frequency',
    'TUE': 'Time_Using_Technology_Daily',
    'CALC': 'Alcohol_Consumption_Frequency',
    'MTRANS': 'Means_of_Transportation',
    'NObeyesdad': 'Obesity_Level'}, axis = 1)
obesity.info()
```

```python
replace_map_fam_history = {
    'yes': 'Yes',
    'no': 'No'
}

obesity['Overweight_Family_History'] = obesity['Overweight_Family_History'].replace(replace_map_fam_history)
obesity['Overweight_Family_History'].value_counts()
```

```python
obesity['High_Calorie_Consumption_Often'].value_counts()
```

```python
replace_map_high_cal = {
    'yes': 'Yes',
    'no': 'No'
}

obesity['High_Calorie_Consumption_Often'] = obesity['High_Calorie_Consumption_Often'].replace(replace_map_high_cal)
obesity['High_Calorie_Consumption_Often'].value_counts()
```

```python
obesity['Vegetable_Consumption_Often'].describe()
```

### Vegetable_Consumption_Often

I rounded the numbers because 810 of them were decimals...I figured this would be the best way to most accurately still include that data without making any crazy assumptions.

```python
obesity['Vegetable_Consumption_Often'] = obesity['Vegetable_Consumption_Often'].round().astype(int)

replace_map_veg = {
    1: 'Never',
    2: 'Sometimes',
    3: 'Always'
}

obesity['Vegetable_Consumption_Often'] = obesity['Vegetable_Consumption_Often'].replace(replace_map_veg)
obesity['Vegetable_Consumption_Often'].value_counts()
```
