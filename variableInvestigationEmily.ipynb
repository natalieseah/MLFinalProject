```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/MIC.csv', index_col=0)
data.head()
```

```python
# top ten variable most correlated with relapse of MI

data.corr(numeric_only=True)[['REC_IM']].sort_values(by='REC_IM', key=abs, ascending=False).head(10)
```

```python
# scatter plots of top 10 - not super helpful because relapse is binary

top10 = ['KFK_BLOOD', 'R_AB_3_n', 'OTEK_LANC', 'NA_R_3_n', 'STENOK_AN', 'NA_R_2_n', 'np_01', 'LET_IS', 'FK_STENOK']
for i in top10:
    plt.scatter(data[i], data['REC_IM'])
    plt.xlabel(i)
    plt.ylabel('REC_IM')
    plt.show()
    #print(data.corr(numeric_only=True)[['REC_IM', i]].sort_values(by='REC_IM', key=abs, ascending=False).head(10))
```
