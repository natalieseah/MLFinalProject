### Fix Column Data Types

```python
unique_values_per_column = {}

for column in mic.columns:
    unique_values_per_column[column] = mic[column].unique()

for column, unique_vals in unique_values_per_column.items():
    print(f"Unique values in '{column}': {unique_vals}")
```

```python
cats = []

for column in mic.columns:
    unique_vals = set(mic[column].dropna().unique())  
    if unique_vals.issubset({0.0, 1.0}):  
        cats.append(column)

print("Categorical columns:", cats)
```

```python
for column in cats:
    mic[column] = mic[column].astype('object')
    
cat_col = [col for col in mic.columns if mic[col].dtype == 'object']
num_col = [col for col in mic.columns if mic[col].dtype != 'object']

print('Categorical columns:', cat_col)
print('Numerical columns:', num_col)
```

### Calculate Missing Values as Percentages

```python
missing_percentage = round((mic.isnull().sum() / mic.shape[0]) * 100, 2)
missing_percentage.sort_values(ascending=False).head(20)
```
